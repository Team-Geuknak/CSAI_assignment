```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "C://Users//tkryu//Desktop//빅데이터 공부//Code States//Section 2//새 폴더//04월 서울시 교통량 조사자료.csv"


df = pd.read_csv(url)

df.head()
```

```python
# 구분, 지점명은 '지점번호'와 같은 의미 이므로 삭제한다.

df2 = df.drop(['구분', '지점명'], axis = 1)
df2.head()
```

```python
# 시간대가 칼럼으로 나눠져있어서 열로 바꿔준다. 
# 시간대를 정렬할 수 있도록 숫자형으로 바꿔준다. 

df3 = df2.melt(id_vars=['일자','요일','지점번호','방향'],
            var_name = '시간',
            value_name = '교통량')
df3['시간'] = df3['시간'].str.replace('시','')
df3['시간'] = pd.to_numeric(df3['시간'], downcast='integer')
df3
```

```python
# 보기 편하도록 정렬해준다.

df3 = df3.sort_values(by = ['일자','지점번호','시간'])
df3.reset_index(inplace=True, drop = True)
```

```python
# 모델을 만들어보니 rows가 너무 많아서 메모리 부족으로 프로그램이 돌아가지 않는다. 
# 때문에 "방향" 구분을 없애고 교통량을 합쳐서 row 개수를 반으로 줄인다. 

df4 = df3.groupby(['일자','요일','지점번호','시간'], as_index=False).agg({'교통량' : 'sum'})
df4.head()
```

```python
# row 개수 확인 
len(df4)
```

```python
# 결측치확인 

df4.isnull().sum()
```

```python
# train / val 나누기

from sklearn.model_selection import train_test_split

train, val = train_test_split(df4, train_size = 0.8, test_size = 0.2, random_state = 2)
```

```python
from sklearn.pipeline import make_pipeline
from category_encoders import TargetEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from category_encoders import CatBoostEncoder
from sklearn.metrics import f1_score
```

```python
target = ['교통량']
features = ['요일','지점번호','시간']

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
```

```python
# 1차 모델을 만들었으나 여전히 메모리가 부족해서 돌아가지 않는다. 
# 


# y_train = np.ravel(y_train)

pipe = make_pipeline(
    CatBoostEncoder(a = 1, sigma = 0),
    SimpleImputer(), 
    RandomForestClassifier(n_jobs=-1, random_state=2, oob_score=True, class_weight='balanced', criterion = "entropy", max_depth = 5)
)

# pipe.fit(X_train,y_train)

print('훈련 정확도: ', pipe.score(X_train, y_train))
print('검증 정확도: ', pipe.score(X_val, y_val))

y_pred_val = pipe.predict(X_val)
y_pred_train = pipe.predict(X_train)
print('훈련 f1 score :',f1_score(y_train, y_pred_train))
print('검증 f1 score :',f1_score(y_val, y_pred_val))
```
