```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "C://Users//tkryu//Desktop//ë¹…ë°ì´í„° ê³µë¶€//Code States//Section 2//ìƒˆ í´ë”//04ì›” ì„œìš¸ì‹œ êµí†µëŸ‰ ì¡°ì‚¬ìë£Œ.csv"


df = pd.read_csv(url)

df.head()
```

```python
# êµ¬ë¶„, ì§€ì ëª…ì€ 'ì§€ì ë²ˆí˜¸'ì™€ ê°™ì€ ì˜ë¯¸ ì´ë¯€ë¡œ ì‚­ì œí•œë‹¤.

df2 = df.drop(['êµ¬ë¶„', 'ì§€ì ëª…'], axis = 1)
df2.head()
```

```python
# ì‹œê°„ëŒ€ê°€ ì¹¼ëŸ¼ìœ¼ë¡œ ë‚˜ëˆ ì ¸ìˆì–´ì„œ ì—´ë¡œ ë°”ê¿”ì¤€ë‹¤. 
# ì‹œê°„ëŒ€ë¥¼ ì •ë ¬í•  ìˆ˜ ìˆë„ë¡ ìˆ«ìí˜•ìœ¼ë¡œ ë°”ê¿”ì¤€ë‹¤. 

df3 = df2.melt(id_vars=['ì¼ì','ìš”ì¼','ì§€ì ë²ˆí˜¸','ë°©í–¥'],
            var_name = 'ì‹œê°„',
            value_name = 'êµí†µëŸ‰')
df3['ì‹œê°„'] = df3['ì‹œê°„'].str.replace('ì‹œ','')
df3['ì‹œê°„'] = pd.to_numeric(df3['ì‹œê°„'], downcast='integer')
df3
```

```python
# ë³´ê¸° í¸í•˜ë„ë¡ ì •ë ¬í•´ì¤€ë‹¤.

df3 = df3.sort_values(by = ['ì¼ì','ì§€ì ë²ˆí˜¸','ì‹œê°„'])
df3.reset_index(inplace=True, drop = True)
```

```python
# ëª¨ë¸ì„ ë§Œë“¤ì–´ë³´ë‹ˆ rowsê°€ ë„ˆë¬´ ë§ì•„ì„œ ë©”ëª¨ë¦¬ ë¶€ì¡±ìœ¼ë¡œ í”„ë¡œê·¸ë¨ì´ ëŒì•„ê°€ì§€ ì•ŠëŠ”ë‹¤. 
# ë•Œë¬¸ì— "ë°©í–¥" êµ¬ë¶„ì„ ì—†ì• ê³  êµí†µëŸ‰ì„ í•©ì³ì„œ row ê°œìˆ˜ë¥¼ ë°˜ìœ¼ë¡œ ì¤„ì¸ë‹¤. 

df4 = df3.groupby(['ì¼ì','ìš”ì¼','ì§€ì ë²ˆí˜¸','ì‹œê°„'], as_index=False).agg({'êµí†µëŸ‰' : 'sum'})
df4.head()
```

```python
# row ê°œìˆ˜ í™•ì¸ 
len(df4)
```

```python
# ê²°ì¸¡ì¹˜í™•ì¸ 

df4.isnull().sum()
```

```python
# train / val ë‚˜ëˆ„ê¸°

from sklearn.model_selection import train_test_split

train, val = train_test_split(df4, train_size = 0.8, test_size = 0.2, random_state = 2)
```

```python
from sklearn.pipeline import make_pipeline
from category_encoders import TargetEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from category_encoders import CatBoostEncoder
from sklearn.metrics import f1_score
```

```python
target = ['êµí†µëŸ‰']
features = ['ìš”ì¼','ì§€ì ë²ˆí˜¸','ì‹œê°„']

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
```

```python
# 1ì°¨ ëª¨ë¸ì„ ë§Œë“¤ì—ˆìœ¼ë‚˜ ì—¬ì „íˆ ë©”ëª¨ë¦¬ê°€ ë¶€ì¡±í•´ì„œ ëŒì•„ê°€ì§€ ì•ŠëŠ”ë‹¤. 
# 


# y_train = np.ravel(y_train)

pipe = make_pipeline(
    CatBoostEncoder(a = 1, sigma = 0),
    SimpleImputer(), 
    RandomForestClassifier(n_jobs=-1, random_state=2, oob_score=True, class_weight='balanced', criterion = "entropy", max_depth = 5)
)

# pipe.fit(X_train,y_train)

print('í›ˆë ¨ ì •í™•ë„: ', pipe.score(X_train, y_train))
print('ê²€ì¦ ì •í™•ë„: ', pipe.score(X_val, y_val))

y_pred_val = pipe.predict(X_val)
y_pred_train = pipe.predict(X_train)
print('í›ˆë ¨ f1 score :',f1_score(y_train, y_pred_train))
print('ê²€ì¦ f1 score :',f1_score(y_val, y_pred_val))
```
