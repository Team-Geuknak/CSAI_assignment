```python
# 기준모델 정확도

df3['units_sold'].value_counts(normalize = True)[baseline]
```

```python
# 값이 1개씩만 있는 특성
### merchange_info_subtitle은 뭔지 더 분석해봐야함 ####
unique_features = ['title_orig', 'merchant_name','mechant_info_subtitile', 'merchant_info_subtitle','merchant_id','product_id']

# 타겟, 피쳐 설정

target = 'units_sold'
exclude = unique_features + [target]
features = df3.columns.difference(exclude, sort=False)
```

```python
# 상관관계 확인

corr = df3[df3.columns.difference(unique_features)].corr()
corr
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(corr)
```

```python
# train / val / test 세트 나누기

from sklearn.model_selection import train_test_split

train_val, test = train_test_split(df3, train_size = 0.8, test_size = 0.2, random_state = 2)

train, val = train_test_split(train_val, train_size = 0.8, test_size = 0.2, random_state = 2)
```

```python
from sklearn.pipeline import make_pipeline
from category_encoders import TargetEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from category_encoders import CatBoostEncoder
from category_encoders import OrdinalEncoder
from sklearn.metrics import f1_score
```

```python
X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
y_test = test[target]
```

```python
# 기본 모델
# 과적합이 많이 되었으나, 일단은 1차로 모델을 만들어 봄.
# 아직 프로젝트에 이 데이터를 쓸 지 확실치 않아서, 전처리, 인코딩 등이 미흡...

pipe = make_pipeline(
    CatBoostEncoder(a = 1, sigma = 0),
    SimpleImputer(), 
    RandomForestClassifier(n_jobs=-1, random_state=2, oob_score=True, class_weight='balanced', criterion = "entropy")
)

pipe.fit(X_train,y_train)

print('훈련 정확도: ', pipe.score(X_train, y_train))
print('검증 정확도: ', pipe.score(X_val, y_val))

y_pred_val = pipe.predict(X_val)
y_pred_train = pipe.predict(X_train)
print('훈련 f1 score :',f1_score(y_train, y_pred_train, average='micro'))
print('검증 f1 score :',f1_score(y_val, y_pred_val, average='micro'))
```

```python
### (제출폼 과제) 이곳에서 과제를 진행해 주세요 ###

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer 

cancer_dataset = load_breast_cancer()

cancer_features = cancer_dataset.data
cancer_target = cancer_dataset.target
cancer = pd.DataFrame(data=cancer_features, columns=cancer_dataset.feature_names)
cancer['target'] = cancer_target

C_X_train, C_X_test, C_y_train, C_y_test = train_test_split(cancer_features, cancer_target, test_size=0.2, random_state=2)
print(C_X_train.shape, C_X_test.shape)
```

```python
cancer
```

```python
pipe3 = make_pipeline(
    CatBoostEncoder(),
    SimpleImputer(),
    XGBClassifier(n_estimators=200
                  , random_state=2
                  , n_jobs=-1
                  , max_depth=6
                  , learning_rate=0.2
                 )
)

pipe3.fit(C_X_train, C_y_train)

print('훈련 정확도: ', pipe3.score(C_X_train, C_y_train))
print('Test 정확도: ', pipe3.score(C_X_test, C_y_test))

C_y_pred_test = pipe3.predict(C_X_test)
C_y_pred_train = pipe3.predict(C_X_train)

print('훈련 f1 score :',f1_score(C_y_train, C_y_pred_train))
print('Test f1 score :',f1_score(C_y_test, C_y_pred_test))
```

```python
### 이곳에서 과제를 진행해 주세요 ### 

# 성능이 좋기로 유명한 catboost를 써보자
# 아직 파라미터에 대해 잘 모르니, xgboost와 동일하게 설정

# 과적합이 XGBoost보다는 약간 줄었으며, 검증세트의 점수가 더 높게 나왔다.


from catboost import CatBoostClassifier

pipe4 = Pipeline([
    ('prep',make_pipeline(CatBoostEncoder(),
                         SimpleImputer())),
    ('cat', CatBoostClassifier(n_estimators = 200
                               , random_state = 2
                               , max_depth = 7
                               , learning_rate = 0.2))
])


pipe4.fit(X_train, y_train);

print('훈련 정확도: ', pipe4.score(X_train, y_train))
print('검증 정확도: ', pipe4.score(X_val, y_val))

y_pred_val4 = pipe4.predict(X_val)
y_pred_train4 = pipe4.predict(X_train)
print('훈련 f1 score :',f1_score(y_train, y_pred_train4, average='micro'))
print('검증 f1 score :',f1_score(y_val, y_pred_val4, average='micro'))
```

```python
# 특성 중요도가 더 치우쳐진 것을 볼 수 있다. 

importances = pd.Series(pipe4.named_steps['cat'].feature_importances_, X_train.columns)

%matplotlib inline
import matplotlib.pyplot as plt

n = 20
plt.figure(figsize=(10,n/2))
plt.title(f'Top {n} features')
importances.sort_values()[-n:].plot.barh();
```

```python
permuter = PermutationImportance(
    pipe4.named_steps['cat'], # model
    scoring='accuracy', # metric
    n_iter=5, # 다른 random seed를 사용하여 5번 반복
    random_state=2
)


X_val_transformed = pipe4.named_steps['prep'].transform(X_val)


permuter.fit(X_val_transformed, y_val);

feature_names = X_val.columns.tolist()

eli5.show_weights(
    permuter, 
    top=None, # top n 지정 가능, None 일 경우 모든 특성 
    feature_names=feature_names # list 형식으로 넣어야 합니다
)
```
