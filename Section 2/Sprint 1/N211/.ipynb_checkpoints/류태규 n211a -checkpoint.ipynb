```python

from sklearn.linear_model import LinearRegression

model = LinearRegression()

feature_grade = ['grade']
target_price = ['price']
X_train_grade = df[feature_grade]
y_train_price = df[target_price]

model.fit(X_train_grade, y_train_price)


# grade Í∞íÏóê Îî∞Î•∏ ÏòàÏ∏° Î™®Îç∏ = ÌååÎûÄ ÏÑ†

X_test_grade = [[x] for x in df['grade']]
y_pred_price = model.predict(X_test_grade)

plt.figure(figsize=(10, 10))

plt.ticklabel_format(style='plain')

plt.scatter(X_train_grade, y_train_price, color = 'orange', alpha = 0.2)
plt.plot(X_test_grade, y_pred_price, color='blue', linewidth = 3)
```

```python
# grade 6Ïù∏ Í≤ΩÏö∞ ÏòàÏ∏° Í∞ÄÍ≤©

X_test_grade6 = [[6]] # grade Í∞í ÏûÖÎ†•
y_pred_grade6 = model.predict(X_test_grade6)

y_pred_grade6
```

# 4


```python
# 'grade' ÌäπÏÑ±Ïóê ÎåÄÌïú ÌöåÍ∑Ä Í≥ÑÏàò coefficient

model.fit(X_train_grade, y_train_price)

print("Í≥ÑÏàò : {}".format(model.coef_))
print("Ï†àÌé∏ : {}".format(model.intercept_))
```

```python
def prediction(grade):
    y_pred = model.predict([[grade]])
    pred = f"{int(grade)} grade ÏòàÏ∏° Í∞ÄÍ≤©: ${int(y_pred[0])} (1 gradeÎãπ Ï∂îÍ∞ÄÍ∏à: ${int(model.coef_[0])})"
    return pred

print(prediction(5))
```

```python
# ÏòàÏ∏° Î™®Îç∏ Í∑∏ÎûòÌîÑ
# ^y = ax + b Í∑∏ÎûòÌîÑ Í∑∏Î†§Î≥¥Í∏∞

x = np.arange(0,15)
y = x * model.coef_[0] + model.intercept_

fig, ax = plt.subplots()

plt.xlim([0,10])
plt.title("Price per Grades prediction model")

ax.plot(x, y, color = 'blue')
ax.ticklabel_format(useOffset=False)
ax.ticklabel_format(style='plain')

plt.show()


# Í≥ÑÏàò coefficient(208457)ÏùÄ Í∏∞Ïö∏Í∏∞Î°ú grade Í∞íÏù¥ 1 Ï¶ùÍ∞Ä Ìï† ÎïåÎßàÎã§ 208457ÎßåÌÅº Í∞ÄÍ≤©Ïù¥ ÏÉÅÏäπÌïúÎã§Îäî Í≤ÉÏùÑ ÎúªÌïúÎã§. 
# Ï†àÌé∏ intercept(-1056044)Îäî gradeÍ∞Ä 0Ïùº Îïå, Í∞ÄÍ≤©Ïù¥ -1058044ÏûÑÏùÑ ÎúªÌïúÎã§. 

from ipywidgets import interact

@interact
def prediction(grade = (1,10)):
    y_pred = model.predict([[grade]])
    pred = f"{int(grade)} grade ÏòàÏ∏° Í∞ÄÍ≤©: ${int(y_pred[0])} (1 gradeÎãπ Ï∂îÍ∞ÄÍ∏à: ${int(model.coef_[0])})"
    return pred
```

```python
### Ïù¥Í≥≥ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥ Ï£ºÏÑ∏Ïöî ###


from sklearn.linear_model import LinearRegression

model = LinearRegression()

feature_sqft = ['sqft_living']
target_price = ['price']
X_train_sqft = df[feature_sqft]
y_train_price = df[target_price]

model.fit(X_train_sqft, y_train_price)



X_test_sqft = [[x] for x in df['sqft_living']]
y_pred_price = model.predict(X_test_sqft)

plt.figure(figsize=(10, 10))

plt.scatter(X_train_sqft, y_train_price, color = 'pink', alpha = 0.5)
plt.plot(X_test_sqft, y_pred_price, color='orange', linewidth = 3)
```

```python
import seaborn as sns

fig, axes = plt.subplots(2, figsize=(10,10))


sns.regplot(x=df['grade'], y=df['price'], ax = axes[0])
sns.regplot(x=df['sqft_living'], y=df['price'], ax = axes[1])


@interact
def prediction(grade = (1,15)):
    y_pred = model.predict([[grade]])
    pred = f"{int(grade)} grade ÏòàÏ∏° Í∞ÄÍ≤©: ${int(y_pred[0])} (1 gradeÎãπ Ï∂îÍ∞ÄÍ∏à: ${int(model.coef_[0])})"
    return pred

@interact
def prediction(sqft = (1,15000)):
    y_pred = model.predict([[sqft]])
    pred = f"{int(sqft)} sqft ÏòàÏ∏° Í∞ÄÍ≤©: ${int(y_pred[0])} (1 sqftÎãπ Ï∂îÍ∞ÄÍ∏à: ${int(model.coef_[0])})"
    return pred
```

```python
from sklearn.metrics import mean_absolute_error

mae_grade = mean_absolute_error(df['price'], X_test_grade)
mae_sqft = mean_absolute_error(df['price'], X_test_sqft)

print("grade Ìï≠Î™©ÏùÑ ÏÇ¨Ïö©Ìïú Í∞ÄÍ≤© ÏòàÏ∏° Î™®Îç∏Ïùò MAE : {}".format(mae_grade))
print("sqft_living Ìï≠Î™©ÏùÑ ÏÇ¨Ïö©Ìïú Í∞ÄÍ≤© ÏòàÏ∏° Î™®Îç∏Ïùò MAE : {}".format(mae_sqft))
print(mae_grade > mae_sqft)
```
