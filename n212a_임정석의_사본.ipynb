```python
# 데이터는 캐글에서 받으셔도 되고 이 링크에서 불러와도 됩니다.
import pandas as pd
df = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kc_house_data/kc_house_data.csv')
```

```python
def convert(date):
  date = date[:4]
  return date

df['date'] = df['date'].apply(convert)
df['date']
```

```python
train = df[df['date']=='2014'].reset_index().drop('index',axis = 1 )
test = df[df['date']=='2015'].reset_index().drop('index',axis = 1 )
```

```python
train
```

```python
test
```

```python
print('훈련데이터의 갯수는: ',len(train))
```

### 2) 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

model = LinearRegression()

features = ['bathrooms','sqft_living']
target = 'price'

X_train = train[features]
X_test = test[features]

y_train = train[target]
y_test = test[target]
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_train)
mae = mean_absolute_error(y_train, y_pred)
print(f'훈련 에러: {mae:.2f}')
```

```python
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print(f'테스트 에러: {mae:.2f}')
```

### 3) 


```python
# 절편과 계수들
#model.intercept_, model.coef_

# 회귀식을 만들어 봅시다.
b0 = model.intercept_
b1, b2 = model.coef_

print(f'y = {b0:.0f} + {b1:.0f}x\u2081 + {b2:.0f}x\u2082')
```
